```python
import os
from PIL import Image
import cv2
path1="H:\\downloads\\notMNIST_small.tar.gz\\notMNIST_small\\notMNIST_small"
F1=os.listdir(path1)
dataName=[]
for i in F1:
    path=path1+"\\"+i
    temp=os.listdir(path)
    dataName.append(temp)
```

```python
n=0
Data=[]
Label=[]
for i in dataName:
    print(F1[n])
    l=[]
    l.append(F1[n])
    temp=[]
    for j in i:
        t=cv2.imread(path1+"\\"+F1[n]+"\\"+j)
        e=[]
        for k in t:
            e1=[]
            for l in k:
                e1.append(l[0])
            e.append(e1)
        temp.append(e)
    Data.append(temp)
    Label.append(l)
    n+=1
```

```python
def f1(l):
    pdt=len(l[0])
    fact=3
    s=0
    D=[]
    c=len(l[0])-2
    for i in l:
        s=0
        s+=sum(i)
        #print(s,pdt)
    
        if(s/pdt<=0.55):
            fact=2
        if(s/pdt<=0.4):
            fact=1
        #print(fact)
        #for i in l:
        x=[]
        for j in range(0,c):
            if(sum(i[j:j+3])>=fact):
                x.append(1)
                continue
            x.append(0)
        D.append(x)
    return D
A=f1(Data[0][12])
for i in A:
    print(i,len(i))
#f1(A)
```

```python
temp=Data[1][12]
l=[]
v=0
for i in temp:
    l1=[]
    for j in i:
        if(j<=50):
            l1.append(0)
            v+=1
            continue
        l1.append(1)
    l.append(l1)
print(v)
for i in range(0,8):
    D=f1(l)
    l = [[D[j][i] for j in range(len(D))] for i in range(len(D[0]))] 
    #print("\n\n\n")
for i in l:
    print(i,len(i))
```
